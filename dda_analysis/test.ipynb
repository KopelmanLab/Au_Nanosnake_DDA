```python
import pandas as pd
import numpy as np
import re

with open('Au_evap') as fp:
    for blank in range(0,3):
        fp.readline()
    wave = []
    Re = []
    Im = []
    for line in fp:
        inter = line.rstrip().split('\t')
        inter = list(chain(*[text.split(' ') for text in inter]))
        ary = [line for line in inter if line]
        
        ary = np.array(ary[0:3]).astype(np.float)
        wave.append(float(ary[0]))
        Re.append(float(ary[1]))
        Im.append(float(ary[2]))


    df = pd.DataFrame({'wave': wave, 'Re': Re, 'Im': Im})

df
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.plot(df['wave'], df['Re'], label="Real")
plt.plot(df['wave'], df['Im'], label="Imaginary")
plt.plot(df['wave'], np.sqrt(df['Re']*df['Re'] + df['Im']*df['Im']), label="Magnitude")

plt.legend()
plt.title('Polarizability vs. Wavelength')
```
